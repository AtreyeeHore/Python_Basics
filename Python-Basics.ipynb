5. Explain the concept of type casting in Python with examples.

Type casting (or type conversion) in Python refers to converting one data type into another. This is useful when you're working with mixed types (e.g., numbers and strings) and want to perform operations that require matching types.

Two Types of Type Casting


i.Implicit Type Casting
Python automatically converts one data type to another when needed.


ii.Explicit Type Casting
The programmer manually converts one data type to another using built-in functions.


```python
a = 5        # int
b = 2.0      # float
c = a + b    # int + float = float

print(c)        # 7.0
print(type(c))  # <class 'float'>
```

```python
x = "10"
y = int(x)    # string to integer
print(y + 5)  # 15
```

6.How do conditional statements work in Python? Illustrate with examples.

Conditional statements allow a Python program to make decisions and execute specific blocks of code depending on whether a condition is True or False.

```python
marks = 75

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Grade: F")
```
